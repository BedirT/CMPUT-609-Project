```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_learning_curve(data):
    p = np.zeros(data.shape[2])
    for i in range(data.shape[2]):
        sm = 0
        for j in range(data.shape[1]):
            sm += data[0][j][i]
        p[i] = sm/data.shape[2]
    plt.plot(p)
```
