# PARAMETER STUDY

```python
s_p_study_1 = np.load('./saves/SARSA-Parameter_Study/rewards.npy')
s_p_study_2 = np.load('./saves/SARSA-Parameter_Study_2-15/rewards.npy')
```

```python
### PARAMETER STUDY FUNC
def parameter_study(data, data2, plot_name):
    ax = plt.subplot()
    num_of_ep = 100

    data = s_p_study_1
    data2 = s_p_study_2

    num_of_params = 10
    num_of_runs = data2.shape[1]
    save = np.zeros(num_of_params)

    for p in range(data.shape[0]):
        res = np.zeros(num_of_ep)
        for e in range(num_of_ep):
            s = 0
            for r in range(data.shape[1]):
                s += data[p][r][e]
            res[e] = s
        save[p] = np.sum(res)

    for p in range(data2.shape[0]):
        res = np.zeros(num_of_ep)
        for e in range(num_of_ep):
            s = 0
            for r in range(data2.shape[1]):
                s += data2[p][r][e]
            res[e] = s
        save[p+5] = np.sum(res)
    print(save.shape)
    ax.plot([-i for i in range(10, 20)], save/num_of_runs)

    # legend = ax.legend(loc='lower right', shadow=True)
    plt.plot()
    plt.savefig(plot_name, transparent=True, dpi=300, quality=100)
```

```python
parameter_study(s_p_study_1, s_p_study_2, 'Sarsa_parameter_study')
```

```python
parameter_study_reinforce_1 = np.load('./saves/Reinforce_Parameter_Study_1/rewards.npy')
parameter_study_reinforce_2 = np.load('./saves/Reinforce_Parameter_Study_2/rewards.npy')
```

```python
parameter_study(parameter_study_reinforce_1, parameter_study_reinforce_2, 'reinforce_parameter_study')
```

# LEARNING CURVES

```python
def plot_learning_curve(data):
    p = np.zeros(data.shape[2])
    for i in range(data.shape[2]):
        sm = 0
#         for j in range(data.shape[1]):
        for j in range(100):            
            sm += data[0][j][i]
        p[i] = sm/data.shape[2]
    return p
```
